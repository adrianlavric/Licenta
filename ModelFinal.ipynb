```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import json
import os
from tensorflow.keras import layers, models, optimizers, callbacks
import warnings
import gc

warnings.filterwarnings('ignore')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
if tf.config.list_physical_devices('GPU'):
    tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)

def eliberare_memorie():
    gc.collect()
    tf.keras.backend.clear_session()
    print("Memoria a fost eliberată manual")

IMG_SIZE = 224
BATCH_SIZE = 16
NUM_EPOCI = 150

SEED = 2025
np.random.seed(SEED)
tf.random.set_seed(SEED)

print('Mediul de lucru:')
print(f'\t• TensorFlow: {tf.__version__}')
print(f'\t• GPU disponibil: {"DA" if tf.config.list_physical_devices("GPU") else "NU"}')

def încarcă_dataset_flori():
    """Încarcă dataset-ul Oxford 102 Flowers"""
    import tensorflow_datasets as tfds

    dataset, info = tfds.load('oxford_flowers102', as_supervised=True, with_info=True)

    train_set = dataset['test']
    val_set = dataset['validation']
    test_set = dataset['train']

    num_train = info.splits['test'].num_examples
    num_val = info.splits['validation'].num_examples
    num_test = info.splits['train'].num_examples
    num_classes = info.features['label'].num_classes

    print('Statistici dataset:')
    print(f'Total clase: {num_classes}')
    print(f'Exemple antrenare: {num_train}')
    print(f'Exemple validare: {num_val}')
    print(f'Exemple testare: {num_test}')

    try:
        with open('label_map.json', 'r') as f:
            class_names = json.load(f)
    except:
        class_names = {str(i): f'Flower_{i}' for i in range(num_classes)}
        with open('label_map.json', 'w') as f:
            json.dump(class_names, f)

    return (train_set, val_set, test_set), info, num_classes, class_names

def normalizare_avansată(imagine, etichetă):
    """Normalizare îmbunătățită cu standardizare"""
    imagine = tf.cast(imagine, tf.float32)
    imagine = tf.image.resize(imagine, [IMG_SIZE, IMG_SIZE])

    imagine = imagine / 255.0

    mean = tf.constant([0.485, 0.456, 0.406])
    std = tf.constant([0.229, 0.224, 0.225])
    imagine = (imagine - mean) / std

    return imagine, etichetă

def augmentare_avansată(img, label):
    """Augmentare mai sofisticata pentru diversitate maxima"""

    if tf.random.uniform(shape=[]) > 0.5:
        img = tf.image.random_flip_left_right(img)

    if tf.random.uniform(shape=[]) > 0.6:
        k = tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32)
        img = tf.image.rot90(img, k=k)

    img = tf.image.random_brightness(img, 0.3)
    img = tf.image.random_contrast(img, 0.7, 1.4)
    img = tf.image.random_saturation(img, 0.7, 1.4)
    img = tf.image.random_hue(img, 0.1)

    if tf.random.uniform(shape=[]) > 0.7:
        scale = tf.random.uniform(shape=[], minval=0.8, maxval=1.2)
        h, w = IMG_SIZE, IMG_SIZE
        new_h = tf.cast(h * scale, tf.int32)
        new_w = tf.cast(w * scale, tf.int32)
        img = tf.image.resize(img, [new_h, new_w])
        img = tf.image.resize_with_crop_or_pad(img, h, w)

    if tf.random.uniform(shape=[]) > 0.8:
        img = tf.image.random_crop(img, size=[int(IMG_SIZE*0.9), int(IMG_SIZE*0.9), 3])
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])

    mean = tf.constant([0.485, 0.456, 0.406])
    std = tf.constant([0.229, 0.224, 0.225])
    img = tf.clip_by_value((img * std + mean), 0.0, 1.0)
    img = (img - mean) / std

    return img, label

def creare_pipeline_optimizat(set_date, batch_size, shuffle_size=2000, este_antrenare=False):
    """Pipeline optimizat cu cache si prefetch imbunatatit"""
    set_date = set_date.map(normalizare_avansată, num_parallel_calls=tf.data.AUTOTUNE)

    if este_antrenare:
        set_date = set_date.cache()
        set_date = set_date.shuffle(shuffle_size, reshuffle_each_iteration=True)
        set_date = set_date.map(augmentare_avansată, num_parallel_calls=tf.data.AUTOTUNE)
        set_date = set_date.repeat(2)

    set_date = set_date.batch(batch_size)
    set_date = set_date.prefetch(tf.data.AUTOTUNE)

    return set_date

def attention_block(inputs, filters):
    """Bloc de atentie pentru focus pe caracteristici importante"""
    gap = layers.GlobalAveragePooling2D()(inputs)
    dense1 = layers.Dense(filters // 8, activation='relu')(gap)
    dense2 = layers.Dense(filters, activation='sigmoid')(dense1)
    channel_attention = layers.Reshape((1, 1, filters))(dense2)

    attended = layers.Multiply()([inputs, channel_attention])

    return attended

def residual_block(x, filters, strides=1):
    """Bloc rezidual pentru gradient flow mai bun"""
    shortcut = x

    x = layers.Conv2D(filters, (3, 3), strides=strides, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)

    x = layers.Conv2D(filters, (3, 3), padding='same')(x)
    x = layers.BatchNormalization()(x)

    if strides != 1 or shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, (1, 1), strides=strides, padding='same')(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)

    x = layers.Add()([x, shortcut])
    x = layers.Activation('relu')(x)

    return x

def creeaza_model_performant_custom(num_classes):
    """
    Model custom de inalta performanta cu attention wi residual blocks
    """
    tf.keras.backend.clear_session()

    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    x = layers.Conv2D(64, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((3, 3), strides=2, padding='same')(x)

    x = residual_block(x, 64)
    x = residual_block(x, 64)
    x = attention_block(x, 64)
    x = layers.Dropout(0.1)(x)

    x = residual_block(x, 128, strides=2)
    x = residual_block(x, 128)
    x = residual_block(x, 128)
    x = attention_block(x, 128)
    x = layers.Dropout(0.2)(x)

    x = residual_block(x, 256, strides=2)
    x = residual_block(x, 256)
    x = residual_block(x, 256)
    x = residual_block(x, 256)
    x = attention_block(x, 256)
    x = layers.Dropout(0.3)(x)

    x = residual_block(x, 512, strides=2)
    x = residual_block(x, 512)
    x = residual_block(x, 512)
    x = attention_block(x, 512)
    x = layers.Dropout(0.4)(x)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(1024, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.6)(x)

    x = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)

    return model

def cosine_annealing_schedule(epoch, lr):
    """Functie pentru cosine annealing learning rate"""
    T_max = NUM_EPOCI
    eta_min = 1e-6
    eta_max = 0.001

    new_lr = eta_min + (eta_max - eta_min) * (1 + np.cos(np.pi * epoch / T_max)) / 2
    return new_lr

def configurare_callbacks_simple_eficient():
    """Callbacks simple si eficiente"""
    model_checkpoint = callbacks.ModelCheckpoint(
        'flori_model_high_performance.keras',
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=False,
        verbose=1
    )

    early_stop = callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=20,
        restore_best_weights=True,
        verbose=1,
        min_delta=0.0005
    )

    reduce_lr = callbacks.ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=8,
        min_lr=1e-7,
        verbose=1,
        mode='max'
    )

    return [model_checkpoint, early_stop, reduce_lr]

def vizualizare_istorie_detaliată(istoric):
    """Vizualizare detaliata a procesului de antrenare"""
    acuratete = istoric.history['accuracy']
    val_acuratete = istoric.history['val_accuracy']
    pierdere = istoric.history['loss']
    val_pierdere = istoric.history['val_loss']
    epoci = range(1, len(acuratete) + 1)

    plt.figure(figsize=(15, 10))

    plt.subplot(2, 2, 1)
    plt.plot(epoci, acuratete, 'ro-', label='Antrenare', linewidth=2, markersize=3)
    plt.plot(epoci, val_acuratete, 'bo-', label='Validare', linewidth=2, markersize=3)
    plt.title('Evolutia Acuratetei', fontsize=16, fontweight='bold')
    plt.xlabel('Epoca', fontsize=12)
    plt.ylabel('Acuratete', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)

    plt.subplot(2, 2, 2)
    plt.plot(epoci, pierdere, 'ro-', label='Antrenare', linewidth=2, markersize=3)
    plt.plot(epoci, val_pierdere, 'bo-', label='Validare', linewidth=2, markersize=3)
    plt.title('Evolutia Pierderii', fontsize=16, fontweight='bold')
    plt.xlabel('Epoca', fontsize=12)
    plt.ylabel('Pierdere', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)

    if 'lr' in istoric.history:
        plt.subplot(2, 2, 3)
        plt.plot(epoci, istoric.history['lr'], 'go-', linewidth=2, markersize=3)
        plt.title('Learning Rate', fontsize=16, fontweight='bold')
        plt.xlabel('Epoca', fontsize=12)
        plt.ylabel('Learning Rate', fontsize=12)
        plt.yscale('log')
        plt.grid(True, linestyle='--', alpha=0.6)
    else:
        plt.subplot(2, 2, 3)
        plt.text(0.5, 0.5, 'Learning Rate\nva fi ajustat automat\nprin ReduceLROnPlateau',
                ha='center', va='center', transform=plt.gca().transAxes,
                fontsize=11, style='italic')

    plt.subplot(2, 2, 4)
    diff_acc = np.array(acuratete) - np.array(val_acuratete)
    diff_loss = np.array(val_pierdere) - np.array(pierdere)
    plt.plot(epoci, diff_acc, 'ro-', label='Diferență Acuratețe', linewidth=2, markersize=3)
    plt.plot(epoci, diff_loss, 'bo-', label='Diferență Pierdere', linewidth=2, markersize=3)
    plt.title('Overfitting Monitor', fontsize=16, fontweight='bold')
    plt.xlabel('Epoca', fontsize=12)
    plt.ylabel('Diferență', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)

    plt.tight_layout()
    plt.show()

    print(f"\n{'='*50}")
    print(f"STATISTICI FINALE DE PERFORMANTA")
    print(f"{'='*50}")
    print(f"Acuratete maxima validare: {max(val_acuratete):.4f} ({max(val_acuratete)*100:.2f}%)")
    print(f"Acuratete finala antrenare: {acuratete[-1]:.4f} ({acuratete[-1]*100:.2f}%)")
    print(f"Acuratete finala validare: {val_acuratete[-1]:.4f} ({val_acuratete[-1]*100:.2f}%)")
    print(f"Pierdere minima validare: {min(val_pierdere):.4f}")
    print(f"Overfitting final: {(acuratete[-1] - val_acuratete[-1])*100:.2f}%")

print("=== INCaRCAREA DATELOR ===")
(train_set, val_set, test_set), dataset_info, num_classes, class_names = încarcă_dataset_flori()

print("\n=== PREGATIREA PIPELINE-URILOR ===")
train_batches = creare_pipeline_optimizat(train_set, BATCH_SIZE, shuffle_size=2000, este_antrenare=True)
val_batches = creare_pipeline_optimizat(val_set, BATCH_SIZE)
test_batches = creare_pipeline_optimizat(test_set, BATCH_SIZE)

eliberare_memorie()

print("\n=== CREAREA MODELULUI ===")
model = creeaza_model_performant_custom(num_classes)
model.summary()

model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-8),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

eliberare_memorie()

print(f"\n=== ANTRENAREA MODELULUI ({NUM_EPOCI} EPOCI) ===")
print("-" * 50)

history = model.fit(
    train_batches,
    epochs=NUM_EPOCI,
    validation_data=val_batches,
    callbacks=configurare_callbacks_simple_eficient(),
    verbose=1
)

vizualizare_istorie_detaliată(history)

print(f"\n{'='*60}")
print("EVALUARE FINALA PE SETUL DE TEST")
print(f"{'='*60}")

test_loss, test_accuracy = model.evaluate(test_batches, verbose=1)
print(f"\n REZULTAT FINAL:")
print(f"   Acuratetea pe test: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"   Loss pe test: {test_loss:.4f}")

model.save('model_flori_high_performance_final.keras')
print(f"\n✅ Modelul a fost salvat ca 'model_flori_high_performance_final.keras'")

eliberare_memorie()

print(f"\n ANTRENAREA COMPLETA!")
print(f"Rezultat final: {test_accuracy*100:.2f}% acuratete pe testare")
```
